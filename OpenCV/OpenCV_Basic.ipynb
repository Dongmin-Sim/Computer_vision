## imread() 확인
* cv2.imread() : 원본 이미지(RGB)를 BGR 형태의 이미지로 반환하여 불러옴  

BGR 형태이기 때문에 불러왔을 때 원하지 않은 색상의, 원본과 다른 색상의 이미지가 불러와지는 것을 확인할 수 있음.

불러온 img 의 `shape`을 확인해보면 2848, 2426, 3 을 확인할 수 있는데 이는 이미지의 사이즈와 채널을 의미함.  
이미지의 `type` 를 확인해보면 `numpy.ndarray` 인 것을 확인할 수 있는데 이를 통해 `imread()` 메서드가 이미지를 넘파이 배열로 불러온 다는 것을 알 수 있음.  
2848은 이미지에서 높이 (행을 의미, 상단부터 0), 2426은 이미지에서 너비 (열을 의미) 가 되고 3은 이미지의 색상을 나타내는 RGB channel 를 의미함

````python
import cv2
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))

# imread('filename') : BGR 로 변환된 이미지
bgr_img = cv2.imread('data/image/walking.jpg')
print('img type : ', type(bgr_img), ' / img shape : ', bgr_img.shape)

plt.imshow(bgr_img)
plt.show()
````

# OpenCV 활용하여 영상 처리
* cv2.VideoCapture : 동영상을 개별 Frame 으로 하나씩 읽어오는 기능
* cv2.Videowriter : VideoCapture 로 읽어들인 Frame 을 동영상으로 다시 write 하는 기능

VideoCapture 를 사용하는 방법은 VideoCapture 객체 생성자를 만들어서 주로 사용함
```python
cap = cv2.VideoCapture('input_video_path', index)
```
- input_video_path : 동영상 경로
- index : 카메라 장치 번호, 0번부터 시작
- cap : VideoCapture 객체

이러한 VideoCapture 객체는 영상, 카메라의 속성을 **확인**하고 **설정**할 수 있는 **get, set** 함수를 제공함  
get(id)에 **속성 아이디**를 전달하면 해당 **입력 동영상 속성의 값**을 알 수 있음  
set(id, value)에 **속성 아이디**와 **값**을 전달하면 **입력 동영상 속성을 지정**할 수 있음.  
```python
# 입력 동영상의 frame 너비 확인
cap.get(cv2.CAP_PROP_FRAME_WIDTH)

# 입력 동영상의 frame 너비(320) 설정
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 320)
```
주요한 'cv2.CAP_PROP_' 속성 ID
- cv2.CAP_PROP_FRAME_WIDTH : 프래임 폭, 너비
- cv2.CAP_PROP_FRAME_HEIGHT : 프래임 높이
- cv2.CAP_PROP_FPS : 초당 프레임 수 


VideoCapture 객체는 read() 함수를 통해 동영상의 영상의 프레임을 읽을 수 있음  
`read()` 함수는 프레임이 정상적으로 읽어졌는지 확인할 수 있는 **Boolean** 과 한개의 프레임 이미지를 표현하는 **Numpy 배열 객체**를 쌍으로 갖는 **튜플을 반환**함.
```python
while True:
    # cap 객체의 프레임을 read() 함수를 통해 차례대로 읽어옴
    # frame 이 정상적으로 읽어지면 read_frame에 True, img_frame 에는 이미지 배열이 들어오게 됨.
    read_frame, img_frame = cap.read()
    # if 조건절을 통해 동영상의 마지막 frame이 끝났을때 반복문을 나올 수 있게
    if not read_frame:
        print('동영상의 frame이 끝났습니다')
        break
```

VideoWrite 를 사용하는 방법은 VideoCapture 와 마찬가지로 VideoWrite 객체생성자를 사용하며
```python
video_writer = cv2.VideoWriter(video_path, fourcc, fps, (width, height))
```
* video_path : 저장할 비디오 경로
* fourcc : Encoding 코덱 유형(4글자), write 시 특정 포맷으로 동영상을 Encoding 가능(DIVX, XVID, MJPG, X264, WMV1, WMV2)
* fps : 초당 프레임수
* (width, height) : 비디오의 너비와 높이(튜플 형식)

위와 같은 인잣값들을 입력받음  

`cv2.VideoWriter_fourcc()`함수 : 4개의 인코딩 포맷문자를 전달하면 코드 값을 생성해내는 함수  
해당 코덱이 있는지 잘 확인하고, encoding 시 설정하는 코덱와 저장하는 동영상의 확장자명을 확인해서 잘 설정해주어야 함.

다음과 같은 형식으로 많이 사용됨.
```python
cap = cv2.VideoCapture(input_video_path)

codec = cv2.VideoWriter_fourcc(*'DIVX')

video_width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
video_height = round(cap.get(cv2.CAP_PROP_FRAM_HEIGHT))

video_fps = cap.get(cv2.CAP_PROP_FPS)

video_writer = cv2.VideoWriter(video_path, codec, video_fps, (video_width, video_height))
```

영상 불러오기 및 VideoWriter 객체 생성

````python
import cv2

# input 비디오 dir
input_video_dir = 'data/video/Trafic.mp4'

# VideoCapture 객체 cap 생성
cap = cv2.VideoCapture(input_video_dir)

# codec 설정 (XVID), (맥os에서 사용, mp4v)
codec = cv2.VideoWriter_fourcc(*"mp4v")

# video 폭, 높이 get 함수로 불러오기
video_width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
video_height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# video fps, get함수로 불러오기
video_fps = cap.get(cv2.CAP_PROP_FPS)

# 맥 Quicktime player에서 재생할 mov 확장자 지정
video_writer = cv2.VideoWriter('data/output/Cars_output.mov', codec, video_fps, (video_width, video_height))

frame_cnt = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
print('frame 개수 : {}, FPS : {}, Frame 크기 : {}'.format(frame_cnt, video_fps, (video_width, video_height)))
````

frame 별로 `read()` 후 frame마다 rectangle 및 caption 적용 후 video_writer에 write

````python
import time

green_color = (0, 255, 0)
red_color = (0, 0, 255)

start = time.time()
index = 0

while True:
    read_frame, img_frame = cap.read()
    if not read_frame:
        print('동영상의 frame 이 끝났습니다.')
        break
    index += 1
    print('frame : ', index, '완료')
    cv2.rectangle(img_frame, (300, 100, 800, 400), color=green_color, thickness=2)
    caption = 'frame : {}'.format(index)
    cv2.putText(img_frame, caption, (300, 95), cv2.FONT_HERSHEY_SIMPLEX, 0.7, red_color, 1)
    
    video_writer.write(img_frame)
    
print('write 완료시간 ', round(time.time()-start, 4))
video_writer.release()
cap.release()    
````
