`imread` 를 사용할 경우 원본이미지의 RGB 값이 아닌 BGR 값으로 읽어오기 때문에 원본사진과 색상이 다름.

```python
import cv2
import matplotlib.pyplot as plt

# imread 로 이미지 불러오기
img = cv2.imread('data/image/red_wall.jpg')
plt.figure(figsize=(10, 10))
plt.imshow(img)
```

cvtColor 를 사용하면 BGR 이미지를 RGB 로 변경할 수 있음. (RGB 뿐만 아니라 grayscale에서 RGB 로 바꾸는 등등 다양한 방법을 인잣값으로 전달하여 변경 가능)

```python
import cv2
import matplotlib.pyplot as plt

bgr_img = cv2.imread('data/image/red_wall.jpg')
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(rgb_img)
```

imwrite() 를 사용하여 BGR 형태의 이미지를 RGB 형태로 변환하여 이미지를 저장할 수 있음

```python
import cv2
import matplotlib.pyplot as plt

bgr_img = cv2.imread('data/image/red_wall.jpg')
cv2.imwrite('data/output/imwrite.jpg', bgr_img)
```
